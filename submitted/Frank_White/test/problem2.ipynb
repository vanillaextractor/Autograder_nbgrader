Consider the following piece of code:

```python
def f(x):
    if x == 0 or x == 1:
        return x
    return f(x - 1) + f(x - 2)
```

---
## Part A (1 point)

Describe, in words, what this code does, and how it does it.

The function computes the Fibonacci sequence recursively.
It returns x for x=0 or x=1.
Otherwise it returns f(x-1) + f(x-2).
